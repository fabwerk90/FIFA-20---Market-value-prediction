```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
```

```python
tm_raw = pd.read_csv("data/tm.csv")
tm_raw["unique_tm_ID"] = tm_raw["player"] + tm_raw["club"] + tm_raw["bday"] 
tm_raw
```

```python
fifa_raw = pd.read_csv("data/fifa.csv")
fifa_raw["unique_fifa_ID2"] = fifa_raw["long_name"] + fifa_raw["dob"]
fifa_raw
```

```python
name_keys = pd.read_csv('data/final_name_keys.csv')
name_keys = name_keys.drop(["Unnamed: 0"],axis=1)
name_keys = name_keys.rename(columns={"player":"player_keys","long_name":"long_name_keys"})

name_keys
```

```python
tm_merge = pd.merge(tm_raw,
                    name_keys,
                    on='unique_tm_ID')

tm_merge = tm_merge.drop_duplicates()
tm_merge["unique_fifa_ID2"] = tm_merge["long_name_keys"] + tm_merge["bday"]
tm_merge
```

# all fine above

```python
fifa_merge = pd.merge(tm_merge,
                      fifa_raw,
                      on='unique_fifa_ID2')
fifa_merge
```

# all fine below

```python
tm_fifa_merge = fifa_merge.copy()
```

```python
# check all columns
pd.set_option('display.max_columns', None)
tm_fifa_merge = tm_fifa_merge.sort_values(by=["overall"],ascending=False)
tm_fifa_merge.head()
```

```python
tm_fifa_merge_cleaned = tm_fifa_merge.copy()
tm_fifa_merge_cleaned = tm_fifa_merge_cleaned.drop(["height_cm_x","unique_tm_ID","player_keys",
                                                    "unique_fifa_ID","sofifa_id","player_url","short_name","age",
                                                    "dob","club_y","value_eur","wage_eur","release_clause_eur",
                                                    "team_position","team_jersey_number","loaned_from","joined",
                                                    "contract_valid_until","nation_position","nation_jersey_number"],
                                                   axis=1)

tm_fifa_merge_cleaned.head()
```

```python
tm_fifa_merge_cleaned = tm_fifa_merge_cleaned
tm_fifa_merge_cleaned = tm_fifa_merge_cleaned.loc[:,~tm_fifa_merge_cleaned.columns.duplicated()]
tm_fifa_merge_cleaned
```
