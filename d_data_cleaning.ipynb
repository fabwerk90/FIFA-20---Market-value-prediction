```python
# Conversion 1
# split dataframe into fieldplayers (fp) and goalkeepers (gk)

# check if every player got a main position
missing_positions = df_raw["main_position"].isnull().sum()
print("Total players missing a main position: {}".format(missing_positions))
df_raw = df_raw.dropna(subset=['main_position'])
missing_positions_afterdrop = df_raw["main_position"].isnull().sum()
print("Total players missing a main position: {}".format(missing_positions_afterdrop))

# split dataframe
df_fp = df_raw.copy()
df_fp = df_fp.loc[df_fp["main_position"] != "Torwart"]

df_gk = df_raw.copy()
df_gk = df_gk.loc[df_gk["main_position"] == "Torwart"]

# reset index
df_fp = df_fp.reset_index(drop=True)
df_gk = df_gk.reset_index(drop=True)
```

```python
# Conversion 2
# drop irrelevant rows for both dataframes

# for field players, these are all goalkeepers attributes as well as the jersey number

df_fp = df_fp.drop(["gk_handling","gk_kicking","gk_reflexes","gk_speed","gk_positioning","gk_diving",
                    "jersey_number","goalkeeping_diving","goalkeeping_handling","goalkeeping_kicking",
                   "goalkeeping_positioning","goalkeeping_reflexes"],axis=1)

# for goalkeepers these are all position-related attributes, all field player attributes and the jersey number
# exceptions here are 
# attacking_short_passing, skill_long_passing, skill_ball_control, movement_acceleration, movement_sprint_speed
# movement_agility, movement_reactions, movement_balance, power_shot_power, power_jumping, power_stamina  
# power_strength, mentality_aggression, mentality_vision, mentality_composure --> Quelle noch rein, warum die für Torwart wichtig sind
df_gk = df_gk.drop(["ls","st","rs","lw","lf","cf","rf","rw","lam","cam","ram","lm","lcm","cm","rcm",
                    "rm","lwb","ldm","cdm","rdm","rwb","lb","lcb","cb","rcb","rb","physic","defending",
                   "dribbling","passing","shooting","pace","jersey_number","attacking_crossing",
                    "attacking_finishing","attacking_heading_accuracy","attacking_volleys","skill_dribbling",
                    "skill_curve","skill_fk_accuracy","power_long_shots","mentality_interceptions",
                    "mentality_positioning","mentality_penalties","defending_marking","defending_standing_tackle",
                    "defending_sliding_tackle"],axis=1)

# reset index
df_fp = df_fp.reset_index(drop=True)
df_gk = df_gk.reset_index(drop=True)
```

```python
# Conversion 3a
# handle missings for field players
print(df_fp.isnull().sum().sort_values(ascending=False))

# apparently there is on entry, which has some inconsistencies
# let's check which index this entry belongs to by arbitrary looking at a missing column
print(df_fp["rb"].isnull().sort_values(ascending=False))

# now that we got the index let's look at this specific entry
df_fp.iloc[10824, :]

# since this player is both considered a defender and goalkeeper let's drop it and reset the index
df_fp = df_fp.drop(10824)

df_fp = df_fp.reset_index(drop=True)

# check again
print(df_fp.isnull().sum().sort_values(ascending=False))
```

```python
# Conversion 3b
# handle missings for goalkeepers
print(df_gk.isnull().sum().sort_values(ascending=False))

# apparently there are 3 entries, which have some inconsistencies
# let's check which indices this entries belong to by arbitrary looking at a missing column
print(df_gk["gk_positioning"].isnull().sort_values(ascending=False))

# now that we got the index let's look at one specific entry
df_gk.iloc[1205, :]

# since these players are both considered a field player and goalkeeper let's drop them and reset the index
df_gk = df_gk.drop([1205, 553, 1063])

df_gk = df_gk.reset_index(drop=True)

# check again
print(df_gk.isnull().sum().sort_values(ascending=False))
```

```python
# Conversion 4
# market value

def mv_conv(df):
    df["number_value"] = df['market_value'].str.split(' ', n=1).str[0]
    df["number_value"] = df["number_value"].apply(lambda x: float(x.split()[0].replace(',', '')))
    df["unit"] = df['market_value'].str.split(' ', n=2).str[1]
    df.loc[df['unit'] == "Mio.", 'multiplicator'] = 10000
    df.loc[df['unit'] == "Tsd.", 'multiplicator'] = 1000
    df["market_value_€"] = df["number_value"] * df["multiplicator"]
    df = df.drop(["market_value","number_value","unit","multiplicator"],axis=1)
    return df

df_fp = mv_conv(df_fp)
df_gk = mv_conv(df_gk)
```

```python
# Conversion 5
# bday

def bday_conv(df):
    
    import datetime

    df["year_of_birth"] = df['bday'].str.split('.').str[2]
    df['bday'] = pd.to_datetime(df['bday'])
    now = datetime.datetime.now()   
    df['player_age'] = ((now - df['bday'])/365).dt.days
    df = df.drop(["bday"],axis=1)
    return df

df_fp = bday_conv(df_fp)
df_gk = bday_conv(df_gk)
```

```python
# Conversion 6
# floats to ints

def float_to_int_conv_fp(df):
    df["divison"] = df["divison"].astype("int64")
    df["national_team_current_or_past"] = df["national_team_current_or_past"].astype("int64")
    df["current_national_player"] = df["current_national_player"].astype("int64")
    df["no_current_national_player"] = df["no_current_national_player"].astype("int64")
    df["national_team_appearances"] = df["national_team_appearances"].astype("int64")
    df["pace"] = df["pace"].astype("int64")
    df["shooting"] = df["shooting"].astype("int64")
    df["passing"] = df["passing"].astype("int64")
    df["dribbling"] = df["dribbling"].astype("int64")
    df["defending"] = df["defending"].astype("int64")
    df["physic"] = df["physic"].astype("int64")
    return df
    
def float_to_int_conv_gk(df):
    df["divison"] = df["divison"].astype(int)
    df["national_team_current_or_past"] = df["national_team_current_or_past"].astype("int64")
    df["current_national_player"] = df["current_national_player"].astype("int64")
    df["no_current_national_player"] = df["no_current_national_player"].astype("int64")
    df["national_team_appearances"] = df["national_team_appearances"].astype("int64")
    df["gk_diving"] = df["gk_diving"].astype("int64")
    df["gk_handling"] = df["gk_handling"].astype("int64")
    df["gk_kicking"] = df["gk_kicking"].astype("int64")
    df["gk_reflexes"] = df["gk_reflexes"].astype("int64")
    df["gk_speed"] = df["gk_speed"].astype("int64")
    df["gk_positioning"] = df["gk_positioning"].astype("int64")
    return df

df_fp = float_to_int_conv_fp(df_fp)
df_gk = float_to_int_conv_gk(df_gk)
```

```python
# Conversion 7
# body_type, real_face

print(df_fp["body_type"].describe())
print("---------------------------------------------------------")
print(df_fp["real_face"].describe())

print("---------------------------------------------------------")
print("---------------------------------------------------------")
print("---------------------------------------------------------")

print(df_gk["body_type"].describe())
print("---------------------------------------------------------")
print(df_gk["real_face"].describe())

# "body_type" yields no explanatory value for us, so let's drop this

df_fp = df_fp.drop(["body_type", "real_face"],axis=1)

# apparently "real_face" is only given for a minority, so let's drop this for now 
# --> can maybe incorporated later for niche-analysis for superstars

df_gk = df_gk.drop(["body_type", "real_face"],axis=1)
```

```python
# Conversion 8
# "+" and "-" out for positions of field players
columns = list(df_fp)

def position_conv_pos(df,column_list):
    for i in column_list:
        if len(i)>3:
            pass
        else:
            df[i] = df[i].str.split('+').str[0]
    return df

def position_conv_neg(df,column_list):
    for i in column_list:
        if len(i)>3:
            pass
        else:
            df[i] = df[i].str.split('-').str[0]
    return df

def position_conv_to_int(df,column_list):
    for i in column_list:
        if len(i)>3:
            pass
        else:
            df[i] = df[i].astype("int64")
    return df
    
df_fp = position_conv_pos(df_fp,columns)
df_fp = position_conv_neg(df_fp,columns)
df_fp = position_conv_to_int(df_fp,columns)
```

```python
# rearrange columns

def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()    
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])

df_fp = movecol(df_fp, cols_to_move=["market_value_€"], ref_col="player_name", place="Before")
df_fp = movecol(df_fp, cols_to_move=["year_of_birth", "player_age"], ref_col="height_cm", place="Before")
df_gk = movecol(df_gk, cols_to_move=["market_value_€"], ref_col="player_name", place="Before")
df_gk = movecol(df_gk, cols_to_move=["year_of_birth", "player_age"], ref_col="height_cm", place="Before")
```
