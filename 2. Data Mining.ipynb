```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
```

```python
tm_raw = pd.read_csv("data/tm.csv")
tm_raw.head(2)
```

```python
fifa_raw = pd.read_csv("data/fifa.csv")
fifa_raw.head(2)
```

```python
clubs = pd.read_csv("data/clubnames.csv")
clubs_dict = pd.Series(clubs.tm.values,index=clubs.fifa).to_dict()
clubs_dict
```

```python
fifa_adj = fifa_raw.copy()
fifa_adj = fifa_adj.replace({"club": clubs_dict})
fifa_adj.head(2)
```

### Build linking criteria
* Link 1 --> Bday + Full Name
* Link 2 --> Birthday + Club
* Link 3 --> Birthday + Height
* Link 4 --> Birthday + First Name
* Link 5 --> Birthday + Last Name

#### Transfermarkt

```python
tm_adj = tm_raw.copy()
tm_adj["ID_iteration_1"] = tm_adj["bday"] + tm_adj["player"]
tm_adj["ID_iteration_2"] = tm_adj["bday"] + tm_adj["club"]
tm_adj["ID_iteration_3"] = tm_adj["bday"] + tm_adj["height_cm"]
tm_adj["ID_iteration_4"] = tm_adj["bday"] + tm_adj["player"].str.split(" ", n=1).str[0]
tm_adj["ID_iteration_5"] = tm_adj["bday"] + tm_adj["player"].str.split(" ", n=1).str[1]
tm_adj
```

#### FIFA

```python
fifa_adj["ID_iteration_1"] = fifa_adj["dob"] + fifa_adj["long_name"]
fifa_adj["ID_iteration_2"] = fifa_adj["dob"] + fifa_adj["club"]
fifa_adj["ID_iteration_3"] = fifa_adj["dob"] + fifa_adj["height_cm"].astype("str")
fifa_adj["ID_iteration_4"] = fifa_adj["dob"] + fifa_adj["long_name"].str.split(" ", n=1).str[0]
fifa_adj["ID_iteration_5"] = fifa_adj["dob"] + fifa_adj["short_name"].str.split(" ", n=1).str[1]
fifa_adj.head(2)
```

```python
tm_merge_iteration_1 = tm_merge[["#","bday","club","height_cm","player","ID_iteration_1"]]
tm_merge_iteration_2 = tm_merge[["#","bday","club","height_cm","player","ID_iteration_2"]]
tm_merge_iteration_3 = tm_merge[["#","bday","club","height_cm","player","ID_iteration_3"]]
tm_merge_iteration_4 = tm_merge[["#","bday","club","height_cm","player","ID_iteration_4"]]
tm_merge_iteration_5 = tm_merge[["#","bday","club","height_cm","player","ID_iteration_5"]]

tm_merge_iteration_1.head()
```

```python
fifa_merge = fifa_adj.copy()
fifa_merge.drop_duplicates(keep = "first", inplace = True)
```

```python
fifa_merge_iteration_1 = fifa_merge[["long_name","dob","height_cm","club","team_jersey_number","ID_iteration_1"]]
fifa_merge_iteration_2 = fifa_merge[["long_name","dob","height_cm","club","team_jersey_number","ID_iteration_2"]]
fifa_merge_iteration_3 = fifa_merge[["long_name","dob","height_cm","club","team_jersey_number","ID_iteration_3"]]
fifa_merge_iteration_4 = fifa_merge[["long_name","dob","height_cm","club","team_jersey_number","ID_iteration_4"]]
fifa_merge_iteration_5 = fifa_merge[["long_name","dob","height_cm","club","team_jersey_number","ID_iteration_5"]]

fifa_merge_iteration_1.head()
```
